```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (assuming you have a DataFrame called 'df' with columns 'iyear' and 'count')
# 'count' can be the number of incidents per year
data = sddata.groupby('iyear').size().reset_index(name='count')

plt.figure(figsize=(10, 6))
plt.plot(data['iyear'], data['count'], marker='o', linestyle='-', color='red')  # Change the line color to red
plt.title('Terrorism Incidents Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (assuming you have a DataFrame called 'df' with columns 'country_txt' and 'count')
# 'count' can be the number of incidents per country
data = sddata['country_txt'].value_counts().reset_index()
data.columns = ['country', 'count']

# Display the top N countries (you can adjust N as needed)
N = 25  # Change this value to display a different number of countries
data = data.head(N)

plt.figure(figsize=(10, 6))
plt.barh(data['country'], data['count'], color='green')  # Change the bar color to green
plt.title('Top {} Countries with Most Terrorism Incidents'.format(N))
plt.xlabel('Number of Incidents')
plt.ylabel('Country')
plt.gca().invert_yaxis()  # Invert the y-axis to display the top countries at the top
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (assuming you have a DataFrame called 'df' with columns 'region_txt' and 'count')
# 'count' can be the number of incidents per region
data = sddata['region_txt'].value_counts().reset_index()
data.columns = ['region', 'count']

plt.figure(figsize=(10, 6))
plt.pie(data['count'], labels=data['region'], autopct='%1.1f%%', startangle=140)
plt.title('Terrorism Incidents by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (assuming you have a DataFrame called 'df' with columns 'region_txt' and 'count')
# 'count' can be the number of incidents per region
data = sddata['region_txt'].value_counts().reset_index()
data.columns = ['region', 'count']

plt.figure(figsize=(10, 6))
plt.bar(data['region'], data['count'], color='violet')  # Change the bar color to green
plt.title('Distribution of Terrorism Incidents by Region')
plt.xlabel('Region')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')  # Rotate and adjust horizontal alignment of x-axis labels
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (assuming you have a DataFrame called 'df' with columns 'region_txt', 'attacktype1_txt', and 'count')
# 'count' can be the number of incidents per combination of region and attack type
data = sddata.groupby(['region_txt', 'attacktype1_txt']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
data.plot(kind='bar', stacked=True)
plt.title('Regional Breakdown of Attack Types')
plt.xlabel('Region')
plt.ylabel('Number of Incidents')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')  # Rotate and adjust horizontal alignment of x-axis labels
plt.tight_layout()
plt.show()
```
